```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)
df['education_man'] = imputer.fit_transform(df[['education_man']]).ravel()
df['num_kids'] = imputer.fit_transform(df[['num_kids']]).ravel()
```

```python
df['divorce_date'] = pd.to_datetime(df['divorce_date'], format='%Y-%m-%d',)
df['div_month'] = df['divorce_date'].dt.month_name().str.slice(stop=3)
df['div_year'] = df['divorce_date'].dt.year
```

```python
df = df.drop('divorce_date',axis=1)
df = df.drop('marriage_date',axis=1)
```

```python
df.head()
```

```python
df['dob_man'] = pd.to_datetime(df['dob_man'], format='%Y-%m-%d',)
df['dob_man_month'] = df['dob_man'].dt.month_name().str.slice(stop=3)
df['dob_man_year'] = df['dob_man'].dt.year

df['dob_woman'] = pd.to_datetime(df['dob_woman'], format='%Y-%m-%d',)
df['dob_woman_month'] = df['dob_woman'].dt.month_name().str.slice(stop=3)
df['dob_woman_year'] = df['dob_woman'].dt.year
```

```python
df.head()
```

```python
df = df.drop('dob_man',axis=1)
df = df.drop('dob_woman',axis=1)
```

```python
df[['education_man', 'education_woman']]=df[['education_man', 'education_woman']].replace('Secondary',0)
df[['education_man', 'education_woman']]=df[['education_man', 'education_woman']].replace('Professional',1)
df[['education_man', 'education_woman']]=df[['education_man', 'education_woman']].replace('Preparatory',2)
df[['education_man', 'education_woman']]=df[['education_man', 'education_woman']].replace('Primary',3)
df[['education_man', 'education_woman']]=df[['education_man', 'education_woman']].replace('Other',4)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
lbl_encoder = LabelEncoder()

# List of columns to encode
columns_to_encode = ['div_month', 'dob_man_month', 'dob_woman_month']

# Encode each column separately
for column in columns_to_encode:
    df[column] = lbl_encoder.fit_transform(df[column])
```
